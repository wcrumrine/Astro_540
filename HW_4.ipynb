```python
#NUenv4 required, using CLASS_InNu-DM (allows fractional case)
#Explores the evolution of TK for f_chi=0.5 for varying redshift

# import necessary modules
import matplotlib
import matplotlib.pyplot as plt
from scipy.special import gamma
from scipy.optimize import fsolve
import numpy as np
from classy import Class
from Helper_Functions1 import Get_LCDM_Pks, Get_WDM_Tx, Calc_K_Max

# for crisper plots
%matplotlib inline
%config InlineBackend.figure_format='retina'
%config InlineBackend.rc = {'figure.facecolor': 'w'}
%config InlineBackend.figure_format = 'retina'
plt.rcParams['font.size']=11
plt.rcParams['text.usetex']=True
plt.rcParams['font.family']='stix'
plt.rcParams['figure.figsize']= (6.5,4)
plt.rcParams['figure.dpi']=150
plt.rcParams['axes.facecolor'] = 'white'
```

```python
# generate Pk's of LCDM model
h=0.6727; omega_b=0.02238280; omega_cdm=0.1201075; tau_reio=0.078; A_s=2.1985e-9; n_s=0.9652; P_k_max=130; k_per_decade=30
k_vec, LCDM_Only = Get_LCDM_Pks(h, omega_b, omega_cdm, tau_reio, A_s, n_s, P_k_max, k_per_decade)
```

```python
# generate Pk's 89% Baryons subbed for DM
h=0.6727; omega_b=0.11*0.02238280; omega_cdm=0.1201075 + 0.89*0.02238280; tau_reio=0.078; A_s=2.1985e-9; n_s=0.9652; P_k_max=130; k_per_decade=30
k_vec, LCDM_Min_Baryons = Get_LCDM_Pks(h, omega_b, omega_cdm, tau_reio, A_s, n_s, P_k_max, k_per_decade)
```
