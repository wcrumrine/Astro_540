```python
#import necessary modules
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
    
# for crisper plots
%matplotlib inline
%config InlineBackend.figure_format='retina'
%config InlineBackend.rc = {'figure.facecolor': 'w'}
```

## Find r(z) for four different cosmological models

```python
####Find r(z) at redshifts ranging 0 to 6 for four different cosmological models
z_vec = np.linspace(0,6,50)

#Define the Integrand used in calculating r(z) for all cosmologies
Integrand = lambda z: ( (Omega_M * (1+z)**3) + (Omega_DM) + (Omega_R * (1+z)**4) + (Omega_k * (1+z)**2) ) ** -0.5

####Einstein de Sitter
Omega_M = 1; Omega_DM = 0; Omega_R = 0; Omega_k = 0
r_z_vec_1 = np.zeros(len(z_vec))

for i in range(len(z_vec)):
    r_z_vec_1[i], err = quad(Integrand, 0, z_vec[i])

    
####Lambda Dominated
Omega_M = 0; Omega_DM = 1; Omega_R = 0; Omega_k = 0
r_z_vec_2 = np.zeros(len(z_vec))

for i in range(len(z_vec)):
    r_z_vec_2[i], err = quad(Integrand, 0, z_vec[i])

    
####LCDM 
Omega_M = 0.3; Omega_DM = 0.7; Omega_R = 0; Omega_k = 0
r_z_vec_3 = np.zeros(len(z_vec))

for i in range(len(z_vec)):
    r_z_vec_3[i], err = quad(Integrand, 0, z_vec[i])
    

####Radiation Dominated
Omega_M = 0; Omega_DM = 0; Omega_R = 1; Omega_k = 0
r_z_vec_4 = np.zeros(len(z_vec))

for i in range(len(z_vec)):
    r_z_vec_4[i], err = quad(Integrand, 0, z_vec[i])


```

## Explore Luminosity Distance vs Redshift

```python
#Calculate Luminosity Distance for redshifts ranging from 0 to 6 for all four cosmologies
D_L_vec_1 = (1+z_vec) * r_z_vec_1
D_L_vec_2 = (1+z_vec) * r_z_vec_2
D_L_vec_3 = (1+z_vec) * r_z_vec_3
D_L_vec_4 = (1+z_vec) * r_z_vec_4
```

```python
#Plot Luminosity vs Redshift for all four models
plt.plot(z_vec, D_L_vec_2, '-c',label=r'Lambda Dominated')
plt.plot(z_vec, D_L_vec_3, '--b',label=r'LCDM')
plt.plot(z_vec, D_L_vec_1, ':k',label=r'Einstein de Sitter')
plt.plot(z_vec, D_L_vec_4, label=r'Radiation Dominated')


plt.xlabel(r'$Redshift$ $(z)$')
plt.ylabel(r'$H_0d_L(z)$')
plt.legend()
plt.title('Luminosity Distance at Observed Redshift')
plt.xlim(0,6)
plt.ylim(0,40)
```

## Explore Angular Diameter Distance vs. Redshift

```python
#Calculate Luminosity Distance for redshifts ranging from 0 to 6 for all four cosmologies
D_A_vec_1 = 1/(1+z_vec) * r_z_vec_1
D_A_vec_2 = 1/(1+z_vec) * r_z_vec_2
D_A_vec_3 = 1/(1+z_vec) * r_z_vec_3
D_A_vec_4 = 1/(1+z_vec) * r_z_vec_4
```

```python
#Plot Luminosity vs Redshift for all four models
plt.plot(z_vec, D_A_vec_2, '-c',label=r'Lambda Dominated')
plt.plot(z_vec, D_A_vec_3, '--b',label=r'LCDM')
plt.plot(z_vec, D_A_vec_1, ':k',label=r'Einstein de Sitter')
plt.plot(z_vec, D_A_vec_4, label=r'Radiation Dominated')


plt.xlabel(r'$Redshift$ $(z)$')
plt.ylabel(r'$H_0d_L(z)$')
plt.legend()
plt.title('Angular Diameter Distance at Observed Redshift')
plt.xlim(0,6)
plt.ylim(0,1.1)
```
